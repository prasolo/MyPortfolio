Higher Dimensions, so we will use Principal Component Analysis to reduce the number of features, in turn reducing the computational time needed to compute the clusters and also to be able to visualise the output.

```python
X = df.iloc[:,5:33]
```

```python
# Creating PCA object
from sklearn.decomposition import PCA
pca = PCA(n_components=2,random_state=42)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca
# 2_D array representing the questions
```

```python
# How much attributed we have retained from the dataset
pca.explained_variance_ratio_.cumsum()[1] # Cumsum to find the cumilative variance
# Hence we retain 87% of Data.
```

```python
pca.explained_variance_ratio_[0]
```

## Model Training

```python
from sklearn.cluster import KMeans
errors = []
for i in range(1,11):
    km = KMeans(n_clusters = i, init = 'k-means++',random_state=42)
    km.fit(X_pca)
    errors.append(km.inertia_)
```

```python
# Elbow Method to find optimum number of clusters
plt.plot (range(1,11),errors,marker ='o',markersize=10)
plt.xticks(np.arange(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Errors')
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, init = 'k-means++',random_state=42)
km.fit(X_pca)
y= km.predict(X_pca)
```

```python
# Trying to Visualize array of X_pca
pd.DataFrame(X_pca)
```

```python
km.cluster_centers_
```

```python
# Plotting the PCA components and clsuter centers predicted on a graph
plt.scatter(X_pca[y==0,0],X_pca[y==0,1],s=50,c='red',label='Cluster 1')
plt.scatter(X_pca[y==1,0],X_pca[y==1,1],s=50,c='yellow',label='Cluster 2')
plt.scatter(X_pca[y==2,0],X_pca[y==2,1],s=50,c='green',label='Cluster 3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100, c='blue',label='Centroids')
plt.title('Cluster of Students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
